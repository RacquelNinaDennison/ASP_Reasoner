```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
df_asp = pd.read_csv("experiments_asp.csv", delimiter="\t")
df_hagen = pd.read_csv("hagen_experiments.csv", delimiter="\t")

# Ensure sorting by 'statements' while keeping corresponding 'total_time' values
df_asp = df_asp.sort_values(by="statements", ascending=True)
df_hagen = df_hagen.sort_values(by="statements", ascending=True)

# Set the theme
sns.set_theme()

# Create the plot
plt.figure(figsize=(10, 6))
sns.lineplot(x=df_asp["statements"], y=df_asp["total_time"], label="ASP Experiments", marker="o")
sns.lineplot(x=df_hagen["statements"], y=df_hagen["total_time"], label="Hagen Experiments", marker="s")

# Labels and title
plt.xlabel("Number of Statements")
plt.ylabel("Total Time")
plt.title("Comparison of Total Time vs Statements")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
